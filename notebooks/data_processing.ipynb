### Join Product dataset with ingredients


````python
import pandas as pd
from rapidfuzz import process, fuzz, distance
import numpy as np
import os
import re
from tqdm import tqdm
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
````

````python
os.listdir('../data/raw/')
````

````python
ingredients_data = pd.read_csv('../data/raw/ingredients_dataset/ingredients_dataset_worldfoods.csv')
products = pd.read_csv('../data/raw/products.csv')
````

#### Stop words Identifier

````python
from collections import Counter

col = products["product_name"].astype(str).str.lower()

# tokenize & clean
tokens = []
for s in col:
    s = re.sub(r"[\W_]+", " ", s)
    tokens.extend(s.split())

# count frequencies
word_counts = Counter(tokens)

# make DataFrame
freq_df = pd.DataFrame(word_counts.items(), columns=["word", "count"]).sort_values("count", ascending=False)

freq_df.head(30)  # view top 30
````

````python
stopwords = {
    "inc", "ltd", "co", "company", "corporation", "corp",
    "pack", "pcs", "pc", "oz", "fl", "ml", "kg", "g", "gram", "grams",
    "ounce", "ounces", "tablet", "bottle", "box", "case", "and", "the",  "100", "with", "of" "for", "a", "an"
    "original"
}
````

#### Clean data and add column

````python
ingredients_data = ingredients_data[['product_name']].drop_duplicates()
products = products[['product_name']].drop_duplicates()
````

````python
products.shape, ingredients_data.shape
````

````python
products["product_name"] = products["product_name"].str.replace("'", "", regex=False)
ingredients_data["product_name"] = ingredients_data["product_name"].str.replace("'", "", regex=False)

products["product_name"] = products["product_name"].str.lower().str.replace(r"[^a-z0-9\s]", " ", regex=True).str.replace(r"\s+", " ", regex=True).str.strip()
ingredients_data["product_name"] = ingredients_data["product_name"].str.lower().str.replace(r"[^a-z0-9\s]", " ", regex=True).str.replace(r"\s+", " ", regex=True).str.strip()
````

````python


def clean_text(s):
    s = str(s).lower()
    s = re.sub(r"[^a-z0-9\s]", " ", s)
    tokens = [t for t in s.split() if t not in stopwords]
    return " ".join(tokens)

products["product_name"] = products["product_name"].map(clean_text)
ingredients_data["product_name"] = ingredients_data["product_name"].map(clean_text)
````

````python
ingredients_data["first_word"] = ingredients_data["product_name"].str.split().str[0]
products["first_word"] = products["product_name"].str.split().str[0]
````

````python
# matches = []
# for i, row in tqdm(products.iterrows(), total=len(products), desc="Fuzzy matching"):
#     subset = ingredients_data[ingredients_data["first_word"] == row["first_word"]]
#     if subset.empty:
#         subset = ingredients_data.sample(300)

#     # compute all four similarity scores for the best candidate group
#     candidates = []
#     for prod in subset["product_name"]:
#         s1 = fuzz.token_set_ratio(row["product_name"], prod)
#         s2 = fuzz.token_sort_ratio(row["product_name"], prod)
#         s3 = fuzz.WRatio(row["product_name"], prod)
#         s4 = distance.JaroWinkler.normalized_similarity(row["product_name"], prod) * 100.0 # Jaro-Winkler
#         avg = (s1 + s2 + s3 + s4) / 4
#         candidates.append((prod, s1, s2, s3, s4, avg))

#     if candidates:
#         best = max(candidates, key=lambda x: x[5])
#         matches.append((
#             row["product_name"],
#             best[0],
#             best[1], best[2], best[3], best[4], best[5]
#         ))


matches = []
for i, row in tqdm(products.iterrows(), total=len(products), desc="Fuzzy matching"):
    subset = ingredients_data[ingredients_data["first_word"] == row["first_word"]]
    if subset.empty:
        subset = ingredients_data.sample(300)

    best_set = process.extractOne(row["product_name"], subset["product_name"], scorer=fuzz.token_set_ratio)
    best_sort = process.extractOne(row["product_name"], subset["product_name"], scorer=fuzz.token_sort_ratio)
    best_wratio = process.extractOne(row["product_name"], subset["product_name"], scorer=fuzz.WRatio)
    best_jaro = process.extractOne(row["product_name"], subset["product_name"], scorer=distance.JaroWinkler.normalized_similarity)  # Jaro-Winkler

    matches.append((
        row["product_name"],
        best_set[0] if best_set else None, best_set[1] if best_set else None,
        best_sort[0] if best_sort else None, best_sort[1] if best_sort else None,
        best_wratio[0] if best_wratio else None, best_wratio[1] if best_wratio else None,
        best_jaro[0] if best_jaro else None, best_jaro[1] if best_jaro else None
    ))


````

````python
matches_df = pd.DataFrame(matches, columns=[
    "product_name",
    "best_match_token_set", "score_token_set",
    "best_match_token_sort", "score_token_sort",
    "best_match_wratio", "score_wratio",
    "best_match_jaro", "score_jaro"
])

````
