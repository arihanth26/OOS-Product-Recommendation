```python

K = 3
dictz = {}
for cluster in data_base.GMM_Cluster_ID.unique():
    print(cluster)
    order_data_sub = combined_order_details_final[combined_order_details_final.product_id.isin(
    data_base[data_base.GMM_Cluster_ID == cluster].Source_P0_Product_ID.unique())]
    
    candidates_data_sub =  data_final[data_final.product_id.isin(
    data_base[data_base.GMM_Cluster_ID == cluster].Source_P0_Product_ID.unique())]

    df_candidates = order_data_sub.merge(
        candidates_data_sub,
        on="product_id",
        how="inner"
    )

    user_sub_purchases = order_data_sub[['product_id', 'order_number', 'order_id', 'user_id' ]].rename(
        columns={"product_id": "substitute_id",
                 "order_number": "sub_order_number"}
    )

    df_join = df_candidates.merge(
        user_sub_purchases,
        on=["user_id", "substitute_id"],
        how="left"
    )

    df_join['future_purchase'] = (
        (df_join['sub_order_number'] > df_join['order_number']) &
        (df_join['sub_order_number'] <= df_join['order_number'] + K)
    )

    labels = (
        df_join.groupby(
            ["user_id", "product_id", "substitute_id", "order_number"]
        )['future_purchase']
        .max()                 
        .reset_index(name='label')
    )

    final_df = df_candidates.merge(
        labels,
        on=["user_id", "product_id", "substitute_id", "order_number"],
        how="left"
    )

    final_df['label'] = final_df['label'].fillna(0).astype(int)
    dictz[cluster] = final_df[['order_id', 'product_id', 'user_id', 'substitute_id', 'label']]
    
```

```python
import pandas as pd
import numpy as np

# ------------------------------------------------------------------------------
# 0. INPUT DATAFRAMES
# sl_final_base: product_id, substitute_id, user_id, order_id, GMM_cluster_id, label
# ing_data:      product-level metadata
# price_data:    product_id, product_name, price_usd
# combined_order_details_final: full Instacart orders
# ------------------------------------------------------------------------------

sl_base = sl_final_base.copy()
ing = ing_data.copy()
price = price_data.copy()
orders = combined_order_details_final.copy()

print("Starting preprocessing...")

# ------------------------------------------------------------------------------
# 1. GLOBAL PRODUCT METADATA (ing + price) WITH NA HANDLING
# ------------------------------------------------------------------------------

print("Step 1: Building global product metadata...")

prod_meta = ing.merge(price[['product_id', 'price_usd']], on='product_id', how='left')

cat_cols = [
    'product_name', 'product_name_cleaned', 'department', 'aisle',
    'brands', 'final_category', 'ingredients_text'
]
for col in cat_cols:
    if col in prod_meta.columns:
        prod_meta[col] = prod_meta[col].fillna('Unknown')

num_cols_meta = [
    'serving_quantity',
    'nutriments_energy-kcal_100g',
    'nutriments_fat_100g',
    'nutriments_sodium_100g',
    'nutriments_carbohydrates_100g',
    'nutriments_salt_100g',
    'nutriments_proteins_100g',
    'nutriments_saturated-fat_100g',
    'nutriments_sugars_100g',
    'price_usd'
]
for col in num_cols_meta:
    if col in prod_meta.columns:
        median_val = prod_meta[col].median()
        prod_meta[col] = prod_meta[col].fillna(median_val)

global_price_median = prod_meta['price_usd'].median()
global_serving_median = prod_meta['serving_quantity'].median()

print("  Global product metadata ready.")

# ------------------------------------------------------------------------------
# 2. GLOBAL PRODUCT-LEVEL STATS FROM ORDERS (POPULARITY, REORDER PROB)
# ------------------------------------------------------------------------------

print("Step 2: Computing global product popularity and reorder stats...")

prod_pop = (
    orders
    .groupby('product_id')
    .size()
    .rename('product_popularity')
    .reset_index()
)

prod_reorder = (
    orders
    .groupby('product_id')['reordered']
    .mean()
    .rename('global_reorder_prob')
    .reset_index()
)

prod_stats = prod_pop.merge(prod_reorder, on='product_id', how='left')

print("  Global product stats ready.")

# ------------------------------------------------------------------------------
# 3. GLOBAL USER-LEVEL STATS
# ------------------------------------------------------------------------------

print("Step 3: Computing global user stats...")

user_total_orders = (
    orders
    .groupby('user_id')['order_number']
    .max()
    .rename('user_total_orders')
    .reset_index()
)

user_reorder_rate = (
    orders
    .groupby('user_id')['reordered']
    .mean()
    .rename('user_reorder_rate')
    .reset_index()
)

user_avg_days = (
    orders
    .groupby('user_id')['days_since_prior_order']
    .mean()
    .rename('user_avg_days_between_orders')
    .reset_index()
)

user_stats = (
    user_total_orders
    .merge(user_reorder_rate, on='user_id', how='left')
    .merge(user_avg_days, on='user_id', how='left')
)

user_stats['user_reorder_rate'] = user_stats['user_reorder_rate'].fillna(0.0)
user_stats['user_avg_days_between_orders'] = user_stats['user_avg_days_between_orders'].fillna(0.0)

print("  Global user stats ready.")

# ------------------------------------------------------------------------------
# 4. GLOBAL ORDER-LEVEL TABLES (BASKET FEATURES + HISTORICAL COUNTS)
# ------------------------------------------------------------------------------

print("Step 4: Preparing global order and basket tables...")

# 4.1 Basket-level features: size and price
print("  Computing basket-level features (size, total price, avg price)...")

orders_with_price = orders.merge(
    price[['product_id', 'price_usd']],
    on='product_id',
    how='left'
)

orders_with_price['price_usd'] = orders_with_price['price_usd'].fillna(global_price_median)

basket_features = (
    orders_with_price
    .groupby(['order_id', 'user_id'])
    .agg(
        basket_size=('product_id', 'size'),
        basket_unique_products=('product_id', 'nunique'),
        basket_total_price=('price_usd', 'sum'),
        basket_avg_price=('price_usd', 'mean')
    )
    .reset_index()
)

print("    basket_features shape:", basket_features.shape)

# 4.2 Order-level time/context (deduplicated per order_id, user_id)
orders_small = (
    orders[['order_id', 'user_id', 'order_number', 'order_dow',
            'order_hour_of_day', 'days_since_prior_order']]
    .drop_duplicates(['order_id', 'user_id'])
)

orders_small = orders_small.merge(
    basket_features,
    on=['order_id', 'user_id'],
    how='left'
)

print("    orders_small shape after merge:", orders_small.shape)
assert not orders_small.duplicated(['order_id', 'user_id']).any()

# 4.3 user-product historical counts (before each order)
print("  Computing global user-product historical frequencies...")

orders_up_prod = (
    orders[['user_id', 'product_id', 'order_number']]
    .sort_values(['user_id', 'product_id', 'order_number'])
)
orders_up_prod['user_product_frequency_before'] = (
    orders_up_prod
    .groupby(['user_id', 'product_id'])
    .cumcount()
)

# 4.4 user-substitute historical counts (treat substitute_id as product_id)
print("  Computing global user-substitute historical frequencies...")

orders_up_sub = (
    orders[['user_id', 'product_id', 'order_number']]
    .rename(columns={'product_id': 'substitute_id'})
    .sort_values(['user_id', 'substitute_id', 'order_number'])
)
orders_up_sub['user_substitute_frequency_before'] = (
    orders_up_sub
    .groupby(['user_id', 'substitute_id'])
    .cumcount()
)

print("  Global order tables ready.")

# ------------------------------------------------------------------------------
# 5. LOOP OVER GMM CLUSTERS
# ------------------------------------------------------------------------------

cluster_ids = sorted(sl_base['GMM_cluster_id'].unique())
print(f"Step 5: Processing {len(cluster_ids)} GMM clusters in a loop...")

cluster_results = []

for cid in cluster_ids:
    print(f"\n=== Processing GMM_cluster_id = {cid} ===")

    sl = sl_base[sl_base['GMM_cluster_id'] == cid].copy()
    print(f"  sl_cluster initial shape: {sl.shape}")

    # 5.1 Merge order-level context + basket features
    print("  Merging order-level context + basket features (orders_small)...")
    sl = sl.merge(orders_small, on=['order_id', 'user_id'], how='left')
    print(f"    After merge orders_small: {sl.shape}")
    assert sl.shape[0] == sl_base[sl_base['GMM_cluster_id'] == cid].shape[0]

    sl['days_since_prior_order'] = sl['days_since_prior_order'].fillna(0.0)

    # 5.2 Merge user-product historical frequency
    print("  Merging user-product historical frequency...")
    sl = sl.merge(
        orders_up_prod,
        on=['user_id', 'product_id', 'order_number'],
        how='left'
    )
    print(f"    After merge user-product freq: {sl.shape}")

    # 5.3 Merge user-substitute historical frequency
    print("  Merging user-substitute historical frequency...")
    sl = sl.merge(
        orders_up_sub,
        on=['user_id', 'substitute_id', 'order_number'],
        how='left'
    )
    print(f"    After merge user-substitute freq: {sl.shape}")

    sl['user_product_frequency_before'] = sl['user_product_frequency_before'].fillna(0).astype(int)
    sl['user_substitute_frequency_before'] = sl['user_substitute_frequency_before'].fillna(0).astype(int)
    sl['user_substitute_past_purchases'] = (sl['user_substitute_frequency_before'] > 0).astype(int)

    # 5.4 Filter product meta/stats to only products in this cluster
    print("  Preparing cluster-specific product metadata and stats...")
    prod_ids_in_cluster = pd.unique(
        pd.concat([sl['product_id'], sl['substitute_id']], ignore_index=True)
    )
    prod_meta_cluster = prod_meta[prod_meta['product_id'].isin(prod_ids_in_cluster)].copy()
    prod_stats_cluster = prod_stats[prod_stats['product_id'].isin(prod_ids_in_cluster)].copy()

    # 5.5 Attach product metadata for original product
    print("  Merging product metadata for original products...")
    prod_meta_prod = prod_meta_cluster.add_prefix('prod_')
    prod_meta_prod = prod_meta_prod.rename(columns={'prod_product_id': 'product_id'})
    sl = sl.merge(prod_meta_prod, on='product_id', how='left')
    print(f"    After merge prod_meta_prod: {sl.shape}")

    # 5.6 Attach product metadata for substitute
    print("  Merging product metadata for substitutes...")
    prod_meta_sub = prod_meta_cluster.add_prefix('sub_')
    prod_meta_sub = prod_meta_sub.rename(columns={'sub_product_id': 'substitute_id'})
    sl = sl.merge(prod_meta_sub, on='substitute_id', how='left')
    print(f"    After merge prod_meta_sub: {sl.shape}")

    # 5.7 Attach product stats (popularity, reorder prob)
    print("  Merging product stats for original products...")
    prod_stats_prod = prod_stats_cluster.add_prefix('prod_')
    prod_stats_prod = prod_stats_prod.rename(columns={'prod_product_id': 'product_id'})
    sl = sl.merge(prod_stats_prod, on='product_id', how='left')
    print(f"    After merge prod_stats_prod: {sl.shape}")

    print("  Merging product stats for substitutes...")
    prod_stats_sub = prod_stats_cluster.add_prefix('sub_')
    prod_stats_sub = prod_stats_sub.rename(columns={'sub_product_id': 'substitute_id'})
    sl = sl.merge(prod_stats_sub, on='substitute_id', how='left')
    print(f"    After merge prod_stats_sub: {sl.shape}")

    # 5.8 Relationship features between product and substitute
    print("  Creating relationship features (same dept/aisle/brand/category, price, serving, nutrition)...")

    sl['same_department'] = (sl['prod_department'] == sl['sub_department']).astype(int)
    sl['same_aisle'] = (sl['prod_aisle'] == sl['sub_aisle']).astype(int)
    sl['same_brand'] = (sl['prod_brands'] == sl['sub_brands']).astype(int)
    sl['same_final_category'] = (sl['prod_final_category'] == sl['sub_final_category']).astype(int)

    sl['prod_price_usd'] = sl['prod_price_usd'].fillna(global_price_median)
    sl['sub_price_usd'] = sl['sub_price_usd'].fillna(global_price_median)
    sl['price_diff'] = sl['sub_price_usd'] - sl['prod_price_usd']
    sl['price_ratio'] = np.where(
        sl['prod_price_usd'] > 0,
        sl['sub_price_usd'] / sl['prod_price_usd'],
        1.0
    )

    sl['prod_serving_quantity'] = sl['prod_serving_quantity'].fillna(global_serving_median)
    sl['sub_serving_quantity'] = sl['sub_serving_quantity'].fillna(global_serving_median)
    sl['serving_quantity_diff'] = sl['sub_serving_quantity'] - sl['prod_serving_quantity']
    sl['serving_quantity_ratio'] = np.where(
        sl['prod_serving_quantity'] > 0,
        sl['sub_serving_quantity'] / sl['prod_serving_quantity'],
        1.0
    )

    nutr_cols = [
        'nutriments_energy-kcal_100g',
        'nutriments_fat_100g',
        'nutriments_carbohydrates_100g',
        'nutriments_proteins_100g',
        'nutriments_sugars_100g',
        'nutriments_sodium_100g',
        'nutriments_salt_100g',
        'nutriments_saturated-fat_100g',
    ]
    for col in nutr_cols:
        prod_col = f'prod_{col}'
        sub_col = f'sub_{col}'
        if prod_col in sl.columns and sub_col in sl.columns:
            diff_col = f'{col}_diff'
            sl[diff_col] = sl[sub_col] - sl[prod_col]

    sl['prod_ingredients_len'] = sl['prod_ingredients_text'].astype(str).str.len()
    sl['sub_ingredients_len'] = sl['sub_ingredients_text'].astype(str).str.len()
    sl['ingredients_len_diff'] = sl['sub_ingredients_len'] - sl['prod_ingredients_len']

    print("    Relationship features created.")

    # 5.9 Merge user stats
    print("  Merging user-level stats...")
    sl = sl.merge(user_stats, on='user_id', how='left')
    print(f"    After merge user_stats: {sl.shape}")

    sl['user_total_orders'] = sl['user_total_orders'].fillna(0).astype(int)
    sl['user_reorder_rate'] = sl['user_reorder_rate'].fillna(0.0)
    sl['user_avg_days_between_orders'] = sl['user_avg_days_between_orders'].fillna(0.0)

    # 5.10 Final numeric NA fill for this cluster
    print("  Final numeric NA fill for this cluster...")
    numeric_cols_all = sl.select_dtypes(include=[np.number]).columns.tolist()
    numeric_cols_all = [c for c in numeric_cols_all if c != 'label']
    for col in numeric_cols_all:
        median_val = sl[col].median()
        sl[col] = sl[col].fillna(median_val)

    # 5.11 Select final modeling columns
    print("  Selecting final columns for this cluster...")

    id_cols = ['user_id', 'order_id', 'product_id', 'substitute_id']

    feature_cols = [
        'GMM_cluster_id',

        'prod_price_usd',
        'sub_price_usd',
        'price_diff',
        'price_ratio',
        'prod_serving_quantity',
        'sub_serving_quantity',
        'serving_quantity_diff',
        'serving_quantity_ratio',

        'prod_product_popularity',
        'sub_product_popularity',
        'prod_global_reorder_prob',
        'sub_global_reorder_prob',

        'same_department',
        'same_aisle',
        'same_brand',
        'same_final_category',

        'nutriments_energy-kcal_100g_diff',
        'nutriments_fat_100g_diff',
        'nutriments_carbohydrates_100g_diff',
        'nutriments_proteins_100g_diff',
        'nutriments_sugars_100g_diff',
        'nutriments_sodium_100g_diff',
        'nutriments_salt_100g_diff',
        'nutriments_saturated-fat_100g_diff',

        'prod_ingredients_len',
        'sub_ingredients_len',
        'ingredients_len_diff',

        'user_total_orders',
        'user_reorder_rate',
        'user_avg_days_between_orders',

        'user_product_frequency_before',
        'user_substitute_frequency_before',
        'user_substitute_past_purchases',

        'order_number',
        'order_dow',
        'order_hour_of_day',
        'days_since_prior_order',

        'basket_size',
        'basket_unique_products',
        'basket_total_price',
        'basket_avg_price',
    ]

    target_col = 'label'

    feature_cols_existing = [c for c in feature_cols if c in sl.columns]

    sl_cluster_final = sl[id_cols + feature_cols_existing + [target_col]].copy()
    print(f"  Final shape for cluster {cid}: {sl_cluster_final.shape}")

    cluster_results.append(sl_cluster_final)

print("\nConcatenating all clusters...")
sl_final_for_model = pd.concat(cluster_results, ignore_index=True)
print("Done. Final dataset shape:", sl_final_for_model.shape)
print(sl_final_for_model.info())
```
